```python
import yfinance as yf
import pandas as pd
from market_climate import market_climate as mc
```

```python
# Miner les donn√©es
vix = yf.Ticker('^VIX').history('max')['Close']
vix.name = 'VIX'
tnx = yf.Ticker('^TNX').history('max')['Close']
tnx.name = 'US 10 Yr'
market_climate_df = pd.concat([vix, tnx], axis=1).dropna()
market_climate_df = market_climate_df.rolling(5).mean().dropna()
market_climate_df = market_climate_df.loc['2010':]
```

```python
mc(df=market_climate_df, x_value="VIX", y_value="US 10 Yr", show=True)
```
